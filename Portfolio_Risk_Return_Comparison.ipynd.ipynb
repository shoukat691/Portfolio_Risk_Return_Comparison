```python
import yfinance as yf
import matplotlib.pyplot as plt

# 1. Download data (Using 'Close' instead of 'Adj Close' to avoid error)
tickers = ['AAPL', 'SPY', 'GLD']
df = yf.download(tickers, start="2014-01-01", end="2024-01-01")['Close']

# 2. Normalize (Base 100)
# Formula: (Current Price / First Price) * 100
norm = (df / df.iloc[0]) * 100

# 3. Plotting with professional colors
plt.figure(figsize=(12, 6))
plt.plot(norm['AAPL'], label='Apple (Tech)', color='blue', linewidth=2)
plt.plot(norm['SPY'], label='S&P 500 (Market)', color='red', linewidth=2)
plt.plot(norm['GLD'], label='Gold (Safe Haven)', color='gold', linewidth=2)

# 4. Professional Styling
plt.title('10-Year Portfolio Growth: Stocks vs Market vs Gold', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Growth Value of $100')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# 5. Show Result
plt.show()
```

```python
# Calculate total growth percentage
total_growth = (norm.iloc[-1] - 100)
print("Total Growth Percentage (Last 10 Years):")
print(total_growth)
```
